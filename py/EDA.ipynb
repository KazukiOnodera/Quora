```python
import pandas as pd
import numpy as np
```

```python
cv = pd.read_csv('../output/cv.csv.gz')
```

```python
def eda(n=1):
    for i in range(n):
        ix = np.random.randint(cv.shape[0])
        q1 = cv.ix[ix, 'question1']
        q2 = cv.ix[ix, 'question2']
        y = cv.ix[ix, 'is_duplicate']
        yhat = cv.ix[ix, 'yhat']
        print(q1)
        print(q2,y,yhat)
        print('='*30)
```

```python
eda(5)
```

```python
cv['delta'] = cv.is_duplicate - cv.yhat
cv.sort_values('delta', inplace=1)
cv.reset_index(drop=1,inplace=1)
```

```python
def topN(n=1):
    for i in range(n):
        ix = i
        q1 = cv.ix[ix, 'question1']
        q2 = cv.ix[ix, 'question2']
        y = cv.ix[ix, 'is_duplicate']
        yhat = cv.ix[ix, 'yhat']
        print(q1)
        print(q2,y,yhat)
        print('='*30)
```
